### **線形回帰モデル**

1. 必要モジュールとデータのインポート

```python
#from モジュール名 import クラス名（もしくは関数名や変数名）

from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
from pandas import DataFrame
import numpy as np
```

```python
#インポートしたデータを確認(data / target / feature_names / DESCR)
print(housing)
```

```python
#DESCR変数の中身を確認
print(housing['DESCR'])
```

```python
#data変数(説明変数)の中身を確認
print(housing['data'])
```

```python
#feature_names変数の中身を確認
#カラム名
print(housing['feature_names'])
```

2. データフレームの作成

```python
#　説明変数らをDataFrameへ変換
df = DataFrame(data=housing.data, columns = housing.feature_names)
# 目的変数をDataFrameへ追加
df['PRICE'] = np.array(housing.target)
# 最初の5行を表示
df.head(5)
```

線形単回帰分析

```python
#カラムを指定してデータを表示
df[['MedInc']].head()
```

```python
# 説明変数
data = df.loc[:, ['MedInc']].values
# 目的変数
target = df.loc[:, 'PRICE'].values
```

```python
## sklearnモジュールからLinearRegressionをインポート
from sklearn.linear_model import LinearRegression
```

```python
## sklearnモジュールからLinearRegressionをインポート
from sklearn.linear_model import LinearRegression
# オブジェクト生成
model = LinearRegression()
#model.get_params()
#model = LinearRegression(fit_intercept = True, normalize = False, copy_X = True,  n_jobs = 1)
```

```python
# fit関数でパラメータ推定
model.fit(data, target)
```

```python
#予測
print('median house age in block group ->3',  model.predict([[3]]))
```

## **回帰係数と切片の確認**

```python
# 単回帰の回帰係数と切片を出力
print('推定された回帰係数: %.3f, 推定された切片 : %.3f' % (model.coef_, model.intercept_))
```

```python
# 重回帰の回帰係数と切片を出力
print(model.coef_)
print(model.intercept_)
```

### **モデルの検証**
## **決定係数**

```python
# train_test_splitをインポート
from sklearn.model_selection import train_test_split
```

```python
# 70%を学習用、30%を検証用データにするよう分割
X_train, X_test, y_train, y_test = train_test_split(data, target,
test_size = 0.3, random_state = 666)
# 学習用データでパラメータ推定
model.fit(X_train, y_train)
# 作成したモデルから予測（学習用、検証用モデル使用）
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
# matplotlibをインポート
import matplotlib.pyplot as plt
# Jupyterを利用していたら、以下のおまじないを書くとnotebook上に図が表示
%matplotlib inline
# 学習用、検証用それぞれで残差をプロット
plt.scatter(y_train_pred, y_train_pred - y_train, c = 'blue', marker = 'o', label = 'Train Data')
plt.scatter(y_test_pred, y_test_pred - y_test, c = 'lightgreen', marker = 's', label = 'Test Data')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
# 凡例を左上に表示
plt.legend(loc = 'upper left')
# y = 0に直線を引く
plt.hlines(y = 0, xmin = -10, xmax = 50, lw = 2, color = 'red')
plt.xlim([0,10 ])
plt.show()
```

```python
# 平均二乗誤差を評価するためのメソッドを呼び出し
from sklearn.metrics import mean_squared_error
# 学習用、検証用データに関して平均二乗誤差を出力
print('MSE Train : %.3f, Test : %.3f' % (mean_squared_error(y_train, y_train_pred), mean_squared_error(y_test, y_test_pred)))
# 学習用、検証用データに関してR^2を出力
print('R^2 Train : %.3f, Test : %.3f' % (model.score(X_train, y_train), model.score(X_test, y_test)))
```
